**Visualization with Plotly**

```python
df_2016 = dt[dt["year"] == 2016]

country_counts = df_2016["country"].value_counts()

# İlk 10 ülkeyi seçme
top_countries = country_counts.head(10)

# Pasta grafiği (Pie Chart) oluşturma
fig = px.pie(
    names=top_countries.index, 
    values=top_countries.values, 
    title="2016 Yılında En Fazla Üniversiteye Sahip 10 Ülke",
    hole=0.3  # Donut görünümü için
)

# Grafiği göster
fig.show()
```

```python
dt = dt.copy()

# convert "num_students" column to numeric
dt["num_students"] = dt["num_students"].astype(str).str.replace(",", "").astype(float)

# convert "total_score" column to numeric
dt["total_score"] = pd.to_numeric(dt["total_score"], errors='coerce')

# Bubble Chart
trace = go.Scatter(
    x=dt["world_rank"],
    y=dt["total_score"],
    mode="markers",
    text=dt["university_name"],
    marker=dict(
        size=dt["num_students"] / 5000,  
        color=dt["total_score"],  
        colorscale="Viridis",
        showscale=True
    )
)

layout = go.Layout(
    title="University Rankings vs. Total Score (Bubble Chart)",
    xaxis=dict(title="World Rank"),
    yaxis=dict(title="Total Score"),
    hovermode="closest"
)

fig = go.Figure(data=[trace], layout=layout)
iplot(fig)
```

```python
trace = go.Box(
    x=dt["country"],
    y=dt["international_students"],
    name="International Students",
    marker=dict(color='maroon')
)

layout = go.Layout(
    title="International Students Percentage by Country",
    xaxis=dict(title="Country"),
    yaxis=dict(title="International Students (%)"),
    hovermode="closest"
)

fig = go.Figure(data=[trace], layout=layout)
iplot(fig)
```

```python
top_10 = dt.sort_values("total_score", ascending=False).head(10)

trace = go.Bar(
    x=top_10["university_name"],
    y=top_10["total_score"],
    marker=dict(color='darkblue')
)

layout = go.Layout(
    title="Top 10 Universities by Total Score",
    xaxis=dict(title="University"),
    yaxis=dict(title="Total Score"),
    hovermode="closest"
)

fig = go.Figure(data=[trace], layout=layout)
iplot(fig)
```

```python
uni_count_per_country = dt["country"].value_counts().reset_index()
uni_count_per_country.columns = ["country", "count"]

trace = go.Bar(
    x=uni_count_per_country["country"],
    y=uni_count_per_country["count"],
    marker=dict(color='lightblue')
)

layout = go.Layout(
    title="Number of Universities per Country",
    xaxis=dict(title="Country"),
    yaxis=dict(title="Number of Universities"),
    hovermode="closest"
)

fig = go.Figure(data=[trace], layout=layout)
iplot(fig)
```
